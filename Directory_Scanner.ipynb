**Folder to scan**

```python
folder_path = "Scotland"
```

**Get all items in the directory**

```python
items = os.listdir(folder_path)
print(items)
```

**Adding "For loop" to visit each item**

```python
for item in items:
  print (item)
```

**Creating Full path**

```python
for item in items:
  full_path = os.path.join(folder_path, item)
  print(full_path)
```

**Indentifying folders and files**

```python
for item in items:
  full_path = os.path.join(folder_path, item)

  if os.path.isfile(full_path):
    print(item, "→ This is a File")

  elif os.path.isdir(full_path):
    print(item, "→ This is a Folder")
```

**Retrieve file Info**

```python
for item in items:
  full_path = os.path.join(folder_path, item)

  if os.path.isfile(full_path):
    size = os.path.getsize(full_path)
    extension = os.path.splitext(item)[1]
    modified = datetime.fromtimestamp(os.path.getmtime(full_path))

    print("File:", item)
    print("Size:", size, "bytes")
    print("Extension:", extension)
    print("Modified:", modified)
    print()

  elif os.path.isdir(full_path):
    print("Folder:", item)
    print()
```

**Saving results to a text file**

```python
report = open("Directory_Report.txt", "w")
for item in items:
  full_path = os.path.join(folder_path, item)
  if os.path.isfile(full_path):
    size = os.path.getsize(full_path)
    extension = os.path.splitext(item)[1]
    modified = datetime.fromtimestamp(os.path.getmtime(full_path))

    report.write(f"File: {item}\n")
    report.write(f"Size: {size} bytes\n")
    report.write(f"Extension: {extension}\n")
    report.write(f"Modified: {modified}\n\n")

  elif os.path.isdir(full_path):
    report.write(f"Folder: {item}\n\n")

report.close()
print("Directory report created successfully!")
```

**Display report**

```python
with open("Directory_Report.txt", "r") as report:
  print(report.read())
```
